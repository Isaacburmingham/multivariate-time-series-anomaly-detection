### LASP Data Exploration

This file converts the original LASP data set from https://github.com/sapols/Satellite-Telemetry-Anomaly-Detection where the data was originally parsed as univariate time series data from an orbiting spacecraft, and joined together to make a multivariate time series dataset. Anomaly detection would be useful in this dataset as sensor failure and sudden environment changes for a satellite are cruical elements for the missions integrity. <br/>

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from keras.preprocessing.sequence import TimeseriesGenerator
from custom_functions import detectTimeCorrelated
```

```python
batTemp = pd.read_csv(r"C:.\Lasp Original Data\BatteryTempCPV6Averaged.csv")
busVolt = pd.read_csv(r"C:.\Lasp Original Data\Bus Voltage (Averaged).csv")
busCurrent = pd.read_csv(r".\Lasp Original Data\TotalSpacecraftBusCurrentSliced.csv")
wheelTemp = pd.read_csv(r"C:.\Lasp Original Data\TRW1MT(averaged).csv")
```

```python
busVolt.rename({busVolt.columns[0]:"Time"},axis=1,inplace=True)
batTemp.rename({batTemp.columns[0]:"Time"},axis=1,inplace=True)
busCurrent.rename({busCurrent.columns[0]:"Time"},axis=1,inplace=True)
wheelTemp.rename({wheelTemp.columns[0]:"Time"},axis=1,inplace=True)
```

```python
batTemp['Time'] = pd.to_datetime(batTemp['Time'])
busVolt['Time'] = pd.to_datetime(busVolt['Time'])
busCurrent['Time'] = pd.to_datetime(busCurrent['Time'])
wheelTemp['Time'] = pd.to_datetime(wheelTemp['Time'])
```

Dropping unnecessary columns and checking the time cadence of the files

```python
busVolt.drop(['min','max','count'],axis=1,inplace=True)
busVolt.rename({busVolt.columns[1]:'Voltage'},axis=1,inplace=True)

wheelTemp.drop(['min','max','count'],axis=1,inplace=True)
wheelTemp.rename({wheelTemp.columns[1]:'wheelTemp'},axis=1,inplace=True)

busCurrent.rename({busCurrent.columns[1]:'Current'},axis=1,inplace=True)

batTemp.rename({batTemp.columns[1]:'Temperature'},axis=1,inplace=True)
```

```python
print("Battery Temperature Cadence:",batTemp['Time'].diff(1)[1])
print("Bus Voltage Cadence:",busVolt['Time'].diff(1)[1])
print("Wheel Temperature Cadence:",wheelTemp['Time'].diff(1)[1])
print("Bus Current Cadence:",busCurrent['Time'].diff(1)[1])
```

```python
dfs = [batTemp, busVolt, wheelTemp, busCurrent]
```

The below reduce allows us to take a list of pandas DataFrames and merge them together in one line

```python
sensdf = reduce(lambda left,right: pd.merge(left,right,on=['Time'],
                                            how='outer'), dfs)
```

```python
sensdf
```

```python
# Given metadata states 0's were treated as NaN
sensdf.replace(0,np.nan,inplace=True)
```

```python
sensdf.isna().sum()
```

```python
sensdf.sort_values(by='Time',inplace=True)
```

```python
sensdf
```

```python
print("Battery Temperature Window:",batTemp['Time'].min(),batTemp['Time'].max())
print("Bus Voltage Window:",busVolt['Time'].min(),busVolt['Time'].max())
print("Wheel Temperature Window:",wheelTemp['Time'].min(),wheelTemp['Time'].max())
print("Bus Current Window:",busCurrent['Time'].min(),busCurrent['Time'].max())
```

Bus Voltage is the only sensor that has data prior to 2/13/04, so filtering to only include window with all data

```python
sensdf = sensdf[sensdf['Time']>="2004-02-13 13:02:30"]
```

Mean resampling the data to a 3H cadence instead of 5 minutes to reduce noise and better understand data. NaNs were imputated with linear interpolation.

```python
sensdf = sensdf.resample('3H',on='Time').mean().interpolate(method='linear')
```

```python
sensdf.reset_index(inplace=True)
```

```python
sensdf.isna().sum()
```

```python
g = sns.PairGrid(sensdf, vars=['Temperature', 'Voltage', 'wheelTemp', 'Current'])
g.map(plt.scatter, alpha=0.8)
```

You can see just from the above grid that there seems to be a decent amount of outliers that stray from the distribution of the data

```python
sns.heatmap(sensdf.iloc[:,1:].corr());
```

Looks like there's a good mix of correlation between features, with some higher correlations between wheel temperature and current. These features should be adequete for a neural network model.

```python
print("Any variables correlated with time?:",detectTimeCorrelated(sensdf,'Time'))
```

No variables correlated with time, so our data looks ready to be processed for modeling.

```python
fig, axs = plt.subplots(4,figsize=(16,10),sharex=True)

axs[0].plot(sensdf['Time'],sensdf['Temperature'],color='#192C87')
axs[1].plot(sensdf['Time'],sensdf['Voltage'], color='#192C87')
axs[2].plot(sensdf['Time'],sensdf['Current'], color='#192C87')
axs[3].plot(sensdf['Time'],sensdf['wheelTemp'], color='#192C87')

plt.suptitle("LASP Sensors")

axs[0].set_ylabel("Temperature")
axs[1].set_ylabel("Voltage")
axs[2].set_ylabel("Current")
axs[3].set_ylabel("Wheel Temperature")

axs[3].set_xlabel("Timestep")

plt.show()
```
